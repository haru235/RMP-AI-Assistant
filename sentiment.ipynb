```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
import json
```

```python
text = "I love the way this professor teaches, but the exams are way too difficult."

# Analyze sentiment
scores = analyzer.polarity_scores(text)

print(scores)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example Data
data = {
    'date': ['2024-08-01', '2024-08-02', '2024-08-03'],
    'review': [
        "The professor is amazing and really helps students understand the material.",
        "The lectures are boring, and I struggled with the assignments.",
        "Great professor, but the exams are too difficult."
    ]
}

df = pd.DataFrame(data)

# Analyze Sentiments
df['sentiment_score'] = df['review'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'])

sentiment_json = df[['date', 'sentiment_score']].to_json(orient='records', date_format='iso')

with open('sentiment_results.json', 'w') as json_file:
    json_file.write(sentiment_json)

# Plotting
plt.plot(df['date'], df['sentiment_score'], marker='o')
plt.title('Sentiment Score Over Time')
plt.xlabel('Date')
plt.ylabel('Sentiment Score')
plt.show()

print(sentiment_json)
```

```python
with open('sentiment_results.json', 'w') as f:
    json.dump(df, f)
```
